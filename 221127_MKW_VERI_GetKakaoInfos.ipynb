```python
import pandas as pd
import os
```

```python
file_list = [x for x in os.listdir(os.getcwd()+'/csv') if '.csv' in x]
file_list
```

```python
for file in file_list:
    if 'ERR' in file or 'INFO' in file:
        continue
    print(f'{file}')
    df = pd.read_csv(f'./csv/{file}', index_col = 0)
    print(f'origin_shape:{df.shape}')
    idf = pd.read_csv(f'./csv/INFO_{file}', index_col = 0)
    print(f'info_shape:{idf.shape}')
    null_ser = (idf.isna().sum()/(len(idf.index)) * 100)
    print(null_ser[null_ser<70],end='\n\n')
```

```python
for x in ['INFO_KAKAO_small_luxury.csv','INFO_KAKAO_birthday.csv']:
    print(x)
    idf = pd.read_csv(f'./csv/{x}', index_col = 0)
    null_ser = (idf.isna().sum()/(len(idf.index)) * 100)
    print(null_ser, end ='\n\n')
```

```python
df = null_ser.to_frame().sort_values(0)
```
