```python
import pandas as pd
import string
import re
```

# Name Cleaning

```python
df = pd.read_csv('./csv/total_data.csv',index_col=0)
df['name'] = df['name'].str.translate({key:32 for key in str.maketrans('','',string.punctuation)})
df['name'] = df['name'].str.replace(r'(\w*\d\w*)','',regex=True)
df['name'] = df['name'].str.replace('x','')
df.head()
```

# Price Cleaning

```python
df['price'] = df['price'].str.replace(',','')
df['price'] = df['price'].str.replace('원','')
df.head()
```

# Material Cleaning

```python
df['material'] = df['material'].str.replace(r'(\w*\d\w*)','',regex=True)
df['material'] = df['material'].str.translate({key:32 for key in str.maketrans('','',string.punctuation)})
df.head()
```

# 상세  추출

```python
df.loc[df['color'].str.contains('상세')==True, 'color'] = float('nan')
df.loc[df['material'].str.contains('상세')==True, 'material'] = float('nan')
df.loc[df['madefrom'].str.contains('상세')==True, 'madefrom'] = float('nan')
```

# Missing Rate Calc.

```python
df.isnull().sum()/len(df)
```

```python
key_list = ['type','material','madefrom','color']

for key in key_list:
    print(f'\n')
    print(df[[key]].value_counts())
```

```python
df = df.reset_index(drop = True)
```

```python
df.to_csv('./csv/CLEANED_CSV.csv')
```
